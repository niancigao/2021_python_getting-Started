```python
import numpy as np
import matplotlib.pyplot as plt

usd = np.array([29.86,29.78,29.695,29.715,29.685,
               29.665,29.67,29.635,29.625,29.665,
               29.665,29.635,29.66,29.655,29.67,
               29.64,29.61,29.585,29.51,29.34])

gold = np.array([1583.1,1615,1651.35,1642.8,1643.8,
                1674.75,1680.25,1701.65,1704.65,1711.1,
                1685.2,1667.3,1684,1678.75,1702.65,
                1718.1,1707.1,1692.2,1698.65,1708.1])

plt.scatter(usd,gold)                          #散佈圖 (x,y) X，Y 資料一定要輸
plt.show                                       #負相關
```

```python
print(np.corrcoef(usd,gold))    #相關矩陣
                                #美金VS美金=1
                                #黃金VS黃金=1
                                #美金VS黃金=-0.71
```

```python
#回歸(regression)模型 : 用公式表現 ?
#簡單線性回歸 : Y = AX + B

print(np.polyfit(usd,gold,1)) #第三參數為 1 (次方數(二元一次方程式))

#求回歸模型

#[A,B]
```

```python
print(np.polyfit(usd,gold,1).round(1))   #[A,B] Y = -245.1 X + 8944.1
```

```python
coef = np.polyfit(usd,gold,1) #A,B矩陣

reg_model = np.poly1d(coef)   #是數字 1 (1維度)

print(reg_model)              #回歸模型(一階)
```

```python
print(reg_model(29.75).round(3))
```

```python
plt.scatter(usd,gold)                         #散佈圖

plt.plot(usd,reg_model(usd),color='orange')   # Y 軸為預測金價

plt.show()


print(reg_model(usd).round(3))
```

```python
#美編  原:6*4吋

plt.figure(figsize=(12,8)) #設成 12*8  
                           #寫在 matplotlib 其他程式碼之前，不然會產生空白圖

    
# matplotlib 的其他程式
plt.scatter(usd,gold)                                             
plt.plot(usd,reg_model(usd),color='orange')   # Y 軸為預測金價

plt.show()
```

```python

tai=np.array([596,593,596,603,603,612,615,618,613,
610,610,608,604,606,612,611,602,600,587,592,592,590])                                       #倒序
tai1 = tai[::-1]


large=np.array([17068.24,17065.97,17122.16,17078.86,17296.90,17415.30,
17541.36,17559.65,17452.52,17518.13,17634.47,17693.13,17764.04,17841.37,
17818.31,17803.54,17666.12,17642.52,17654.19,17369.39,17328.09,17427.76])

print(np.corrcoef(tai1,large))                   #係數矩陣
plt.scatter(tai1,large)                          #散佈圖 (x,y) X，Y 資料一定要輸

coef = (np.polyfit(tai1,large,1).round(1))       #A，B矩陣
reg_model = np.poly1d(coef)                      #方程式 [[是數字 1 (1維度)
print(reg_model)      

plt.figure(figsize=(12,8))      #圖大小
plt.rcParams['font.size'] = 16  #文字大小
plt.rcParams['font.family'] = 'Microsoft JhengHei'  #(微軟正黑體) 

# matplotlib 的其他程式
plt.scatter(tai1,large)                                             
plt.plot(tai1,reg_model(tai1),color='orange',label='y = 25.4 x + 2189') 
# 折線圖本身的圖例說明

plt.title('回歸分析')
plt.xlabel('台積電11月收盤')
plt.ylabel('大盤收盤價')             #轉套成中文

plt.legend()     #圖例
plt.grid(True)   #格線

plt.show()
```
