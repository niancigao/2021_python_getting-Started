```python
import numpy as np
import matplotlib.pyplot as plt

stock=np.array([271.5,275.5,283,285,283,279.5,278.5,285,287.5,286.5,306.5,304,295,294,295.5,294,298,296.5,299,304.5])

stock2=np.array([311.0,317.5,320.5,323.0,315.0,305.5,307.0,302.0,294.0,290.0,276.5,268.0,260.0,248.0,270.0,255.0,267.5,277.0,280.0,273.0,267.5,274.0])

plt.plot(stock)  #第一條是藍色
plt.plot(stock2)

plt.show()
```

```python
plt.subplot(211)  #先指定設定位置的子圖表
plt.plot(stock)   #在於該子圖表中繪圖
plt.subplot(212)
plt.plot(stock2)
plt.show
```

```python
# plt.subplot(211) :劃分子圖表
# 2 :縱軸數量 ； 1 :橫軸數量 ； 1 :要繪製的第1個子圖表
# 2*1 矩陣 + 順序
```

```python
plt.subplot(221)
plt.plot([1,2,3,4,5])

plt.subplot(222)            #第二張往右
plt.plot([5,4,3,2,1])       #先row再colum

plt.subplot(223)
plt.plot([1,2,4,7,11])

plt.show()
```

```python
plt.subplot(231)
plt.plot([1,2,3,4,5])

plt.subplot(232)            #第二張往右
plt.plot([5,4,3,2,1])       #先row再colum

plt.subplot(233)
plt.plot([1,2,4,7,11])

plt.subplot(234)
plt.plot([1,2,4,7,11,5,5,5,5])

plt.subplot(235)
plt.plot([1,2,4,7,11,5,5,5,5,1,1,1,1,1,1])

plt.show()
```

```python
#資料偏度 (skewnss)
#離散程度 (measures od dispersion)
```

```python
stock=np.array([271.5,275.5,283,285,283,279.5,278.5,285,287.5,
                286.5,306.5,304,295,294,295.5,294,298,296.5,299,304.5])

plt.hist(stock)   #histogram:直方圖
plt.show()
```

```python
plt.hist([stock,stock2])   #用串列方法把兩個陣列傳給 plt.hist()
                           #折線圖(plot)不行
plt.show()
```

```python
plt.hist([stock,stock2],bins=20)   # bins 自訂分組量
                           
plt.show()
```

```python
# box plot(箱型圖)
#右 正偏態 
#左 負偏態

plt.boxplot([stock,stock2])   #把多筆資料放在串列中
                           
plt.show()
```
