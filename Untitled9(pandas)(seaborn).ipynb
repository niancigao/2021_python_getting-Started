```python
#seaborn 有個功能可一次畫兩張圖稱 pair plot

from sklearn import datasets 
import pandas as pd
import matplotlib.pyplot as plt  # matplotlib 大於 seaborn 所以要先打
import seaborn as sns            #而且是一定要打

col_names = ['sepal width','sepal length','petal width','petal length']

df = pd.DataFrame(datasets.load_iris().data,columns = col_names)

sns.pairplot(df)   #兩兩相比較的關係圖 pair plot
plt.show()         #因為 seaborn 是建構在 matplotlib 之上
                   #所以還是要用 plt 產生最終圖表 

#對角線為資料本身直方圖
```

```python
df_cols=df[['sepal width','petal width']]
#篩選的欄位 包成串列再放入 df[]
sns.pairplot(df_cols)
plt.show()
```

```python
df_cols=df[['sepal width','petal width']]

sns.pairplot(df_cols,kind='reg')    #回歸縮寫
plt.show()
```

kind：用於控制非對角線上的圖的類型，可選"scatter"與"reg"

diag_kind：控制對角線上的圖的類型，可選"hist"與"kde(核密度估計(折線圖))" 

```python
sns.set(font_scale=1.5)     #文字放大1.5倍
sns.pairplot(df,height=4)   #高度 4 英吋
```

```python
print(df.corr())    #求相關係數矩陣 by pandas

sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1) #熱圖(heat map) 
            #相關矩陣當資料、將數字資料顯示在圖上、給範圍 (必打最大最小值)
            #否則變邊界最大最小值
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True,vmin=-1,vmax=1,cmap='viridis')  #可換顏色
#cmap(colormap)色系名稱 (viridis:內建)
```

```python
#可挑特定欄位畫熱圖(pair plot 也可)

cols=['sepal width','sepal length']
sns.heatmap(df[cols].corr(),annot=True,vmin=-1,vmax=1)
plt.show()
```

```python
#跟matplotlib比比較接近

sns.set(font_scale=1.5)           #字體1.5倍
plt.figure(figsize=(6,6))         #6*6吋
sns.heatmap(df.corr(),annot=True) #邊界圖
```

```python
#seaborn可畫箱型圖

sns.boxplot(data=df)   #資料要用 data 屬性傳給 sns.boxplot()
plt.show()
```
