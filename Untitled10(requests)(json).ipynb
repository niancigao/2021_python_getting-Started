```python
import pprint

url = 'http://worldtimeapi.org/api/ip'
data = requests.get(url).json()       
#將查詢和解讀JSON的動作合併成一行
           #字典

pprint.pprint(data)
```

```python
#• 看執行結果前,特別說明:
#• 在網路上,網路服務也常被稱為API(Application Programming Interface 應用程式介面)。
#• 意思就是在概念上把這些網址當成網路版的程式函式。
#• 也就是說你的程式不只能呼叫編輯器所安裝套件的函式
#  更能透過網路存取其他人撰寫的函式、取得世界各地有用的資料。
```

```python
import requests,time #一次匯入兩模組

url='http://worldtimeapi.org/api/ip'

while True:                        #無限迴圈
    data=requests.get(url).json()
    print(data['datetime'])        #字典以鍵取值
    time.sleep(5)                  #隔五秒
```

```python
import requests,time

url='http://worldtimeapi.org/api/ip'

for i in range(10):
    data=requests.get(url).json()
    print(data['datetime'])
    time.sleep(5)
```

```python
response=requests.get(url)

if response.status_code == requests.codes.ok:   #判斷服務是否正常回應
    data = response.json()                      #變字典
    print(data['datetime'])
    
else:
    print('網路服務查詢失敗')
```

```python
# 在查詢服務時使用參數
#• 也有許多網路服務在存取它時需要提供一些參數。
#• 譬如以下可查詢特定經緯度位置當天的日出與日落時間案例。

url = 'http://api.sunrise-sunset.org/json?lat=22.75773&lng=121.166549'
                                          #台東市經緯度
data=requests.get(url).json()             #國際標準時間
pprint.pprint(data)
```

```python
# 給網址參數帶入變數
#• 若覺得手動帶入參數很麻煩,可以把參數放進變數,再以f-string字串格式將參數放進網址

latitude = 29.873356578065597
longitude = 31.21404842651381

#字串格式
url = f'http://api.sunrise-sunset.org/json?lat={latitude}&lng={longitude}'

data=requests.get(url).json()           
pprint.pprint(data)

print('\n')
print( f'http://api.sunrise-sunset.org/json?lat={latitude}&lng={longitude}')
     # f-string : f' {} '
```

```python
url = 'https://opendata.cwb.gov.tw/api/v1/rest/datastore/F-C0032-001?Authorization=rdec-key-123-45678-011121314&locationName=%E9%80%A3%E6%B1%9F%E7%B8%A3'
                                                                                                               #連江縣轉碼結果(可轉回中文)
data=requests.get(url).json()           
pprint.pprint(data)
```

```python
#• 鍵location底下是個串列(包括中括號[ ]內),裡面包含了一個以上地點的天氣預報資料。
#• 由於前面我們只選了一個地點,因此這串列只會有單獨一個元素(索引0)。
```

```python
data['records']['location'][0]['weatherElement'][1]['time']
                         #地點索引為 0           降雨機率資料在索引 1 中的 time 鍵
```

```python
pop = data['records']['location'][0]['weatherElement'][1]['time']

for p in pop:                                               #走訪串列(3元素)
    print('預報區間:',p['startTime'],'~',p['endTime'])
    print('降雨機率:',p['parameter']['parameterName']+'%')  #字串可直接加
    print()                                                #多空一行
```

```python
#• 鍵取出的降雨機率值其實是字串(從JSON取得的資料常常不會剛好合用)。
#• 此處我們只是在後面加一個百分比符號,這樣沒什麼關係,但若要將這資料拿來做運算
#  就要記得先用int( )或float( )將它轉換成數值。
```

```python
#網路資料圖形化:以地震震度統計為例


url='https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_week.geojson'

data=requests.get(url).json()

pprint.pprint(data) #排列整齊

#print(data)        #亂亂排
```

```python
quakes = data['features']            #取出包含地震資料的串列
mag = quakes[0]['properties']['mag'] #取出第一筆地震資料的規模

print(mag)
```

```python
quakes = data['features']
mag_label = ['未滿3級','3-4級','4-5級','5-6級','六級以上']
#震度的"標籤"

mag_list = [0,0,0,0,0]
#震度群組數量的串列

for q in quakes:
    mag = q['properties']['mag']
    
    if mag>=6:
        mag_list[4] += 1
    elif mag>=5:
        mag_list[3] += 1
    elif mag>=4:
        mag_list[2] += 1
    elif mag>=3:
        mag_list[1] += 1
    else:
        mag_list[0] += 1
        
print(mag_list)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ['Microsoft JhengHei']
#設定使用中文
plt.bar(mag_label,mag_list) #繪製長條圖
plt.show()
plt.pie(mag_list,labels=mag_label,autopct='%1.1f%%')
#圓餅圖的資料在前 標籤在後 加上百分比標示
plt.show()
```
