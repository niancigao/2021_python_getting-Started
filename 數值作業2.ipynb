```python
from scipy.stats import *
from scipy.stats.mstats import mquantiles
import matplotlib.pyplot as plt
import numpy as np
n = 100
X = gamma.rvs(a=2,scale=0.5, size=100)
```

```python
# Find percentile of X
q = np.linspace(start=0,stop=0.9,num=10)
Q = np.linspace(start=0,stop=0.9,num=100)
plt.figure(figsize=(10,10))

x=gamma.ppf(Q,a=2,scale=0.5)
print(x)


plt.subplot(221)
plt.boxplot(x)

plt.subplot(222)
plt.boxplot(X)

percentileXX = mquantiles(x,prob=q)
percentileX = mquantiles(X,prob=q)



print("percentileXX:",print('\n'),percentileXX)
print("percentileX:",print('\n'),percentileX)

plt.subplot(223)
plt.plot(percentileXX)

plt.subplot(224)
plt.plot(percentileX)

plt.show()
```

```python
# Plot the histogram of X
plt.figure(1)
plt.subplot(121)
plt.hist(x)
plt.subplot(122)
plt.hist(X)
plt.show()
```

```python
# Plot the density function of X
plt.figure(2)
density = kde.gaussian_kde(X)
xs = np.linspace(-1,10,200)
plt.plot(xs,density(xs))
plt.show()
```

```python
# 1, 5, 10  50  ()
count = 0
total = 0
w = [0,0,0]
while True:
    total = w[0]+w[1]*5+w[2]*10
    if total>50:
        w[0] = 0
        break
    elif total==50:
        count += 1
        print(w)
        w[0] = 0
        break

    while True:
        total = w[0]+w[1]*5+w[2]*10
        if total>50:
            w[1] = 0
            break
        elif total==50:
            count += 1
            print(w)
            w[1] = 0
            break

        while True:
            total = w[0]+w[1]*5+w[2]*10
            if total>50:
                w[2] = 0
                break
            elif total==50:
                count += 1
                print(w)
                w[2] = 0
                break
            w[2] += 1

        w[1] += 1

    w[0] += 1

print(count)
```
